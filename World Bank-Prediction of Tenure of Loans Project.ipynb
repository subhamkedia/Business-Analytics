```python
import os
import sys
import scipy as sp
import pandas as pd # dataframes
import numpy as np # algebra & calculus
import datetime # for date time

import matplotlib.pyplot as plt # plotting
import seaborn as sns # plotting

from functools import partial # to reduce df memory consumption by applying to_numeric

import sklearn 

color = sns.color_palette() # adjusting plotting style
import warnings
warnings.filterwarnings('ignore') # silence annoying warnings

%matplotlib inline
```

```python
mdata = pd.read_csv("master.csv")
sdata = pd.read_csv("supplement.csv")
```

```python
country = mdata['Country'].unique()
mdata["Board Approval Date"] = pd.to_datetime(mdata["Board Approval Date"], format = '%m/%d/%Y %H:%M:%S %p' )
mdata['Year'] = mdata["Board Approval Date"].dt.year
country = sorted(country)
```

```python
dic = {}
```

```python
mdata.head()
```

```python
for key in country:
    lis = []
    for i in range(0, mdata.shape[0]):
        if (mdata['Country'][i] == key):
            if (mdata['Year'][i] not in lis):
                lis.append(mdata['Year'][i])
                lis = sorted(lis)
    dic[key] = lis
```

```python
for k in dic:
    for val in dic[k]:
        if (val == 2018):
            dic[k].pop()
dic
```

```python
set1 = set(sdata['Country Name'].unique())
set2 = set(mdata['Country'].unique())
set3 = set1 & set2
temp_lis = list(set3)
temp_lis = sorted(temp_lis)

for key in list(dic):
    if (key not in temp_lis):
        dic.pop(key, None)
```

```python
new = pd.DataFrame(columns=['Country','Year'])
```

```python
for key in dic:
    for values in dic[key]:
        df = pd.DataFrame([[key, values]], columns=['Country','Year'])
        new = new.append(df)
```

```python
lis2 = sdata['Series Name'].unique()
lis2 = lis2[:-1]
```

```python
for i in lis2:
    new[i] = np.nan
```

```python
new = new.reset_index(drop=True)
```

```python
for country in dic:
    for year in dic[country]:
        for i in lis2:
            data = pd.DataFrame(sdata[sdata['Country Name'] ==  country])
            data = data[data['Series Name'] == i]
            
            temp = pd.DataFrame(new[new['Country'] == country])
            temp = temp[temp['Year']  == year]
            ind = temp.index.values[0]

            new[i][ind] = data[str(year)].iloc[0]
```
